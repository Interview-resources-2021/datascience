### Color Coded Plots

```python
iris_set_data = pd.read_csv('..\datasets\iris_setosa.csv', index_col='Id')
iris_ver_data = pd.read_csv('..\datasets\iris_versicolor.csv', index_col='Id')
iris_vir_data = pd.read_csv('..\datasets\iris_virginica.csv', index_col='Id')

iris_ver_data.head()
```

```python
plt.figure(figsize=(10, 6))
sns.distplot(a=iris_set_data['Petal Length (cm)'], kde=False, label='Iris-setosa', color='y')
sns.distplot(a=iris_ver_data['Petal Length (cm)'], kde=False, label='Iris-versicolor', color='g')
sns.distplot(a=iris_vir_data['Petal Length (cm)'], kde=False, label='Iris-virginica', color='r')

plt.title('Histogram of Petal Lengths by species')
plt.legend()
```

### Multiple KDE Plots

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(data=iris_set_data['Petal Length (cm)'], shade=True, label='Iris-setosa')
sns.kdeplot(data=iris_ver_data['Petal Length (cm)'], shade=True, label='Iris-versicolor')
sns.kdeplot(data=iris_vir_data['Petal Length (cm)'], shade=True, label='Iris-virginica')

plt.title('Distribution of Petal lengths by species')
```
