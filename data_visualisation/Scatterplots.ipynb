```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
candy_data = pd.read_csv('../datasets/candy.csv', index_col='id')
```

```python
candy_data.head()
```

```python
sns.scatterplot(x=candy_data['sugarpercent'], y=candy_data['winpercent'])
```

```python
sns.regplot(x=candy_data['sugarpercent'], y=candy_data['winpercent'], )
```

```python
# Scatter plot showing the relationship between 'pricepercent', 'winpercent', and 'chocolate'
sns.scatterplot(x=candy_data['pricepercent'], y=candy_data['winpercent'], hue=candy_data['chocolate']) # Your code here
```

```python
sns.lmplot(x='pricepercent', y='winpercent',
           data=candy_data, hue='chocolate')
```

We observe that 
- expensive chocolate candies are more popular than cheaper chocolate candies
- cheaper candies w/o chocolate have higher popularity than expensive candies with chocolate

```python
sns.swarmplot(x='chocolate', y='winpercent', data=candy_data)
```
