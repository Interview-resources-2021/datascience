```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
flight_data = pd.read_csv('../datasets/flight_delays.csv', index_col='Month')
```

```python
flight_data.tail()
```

Bar Plot 

```python
plt.figure(figsize=(16, 6))
plt.title('Average Arrival delay for Spirit Airlines Flights, by Month')
sns.barplot(x=flight_data.index, y=flight_data['NK'])
plt.ylabel('Arrival delay in minutes')
```

Heat Map

```python
plt.figure(figsize=(14, 7))
plt.title("Average Arrival Delay for each airline by month")
sns.heatmap(data=flight_data, annot=True)
plt.xlabel('Airline')
```

```python
plt.figure(figsize=(16, 6))
sns.catplot(data=flight_data, aspect=2, kind='swarm')
plt.ylabel('Delay in minutes')
plt.xlabel('Airline')
```

```python
plt.figure(figsize=(16, 6))
sns.boxenplot(data=flight_data)
```

```python
plt.figure(figsize=(16, 6))
sns.violinplot(data=flight_data, inner='stick', palette='pastel', split=True)
```
