```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
museum_data = pd.read_csv('../datasets/museum_visitors.csv', 
                          index_col='Date', parse_dates=True)
```

```python
museum_data.tail()
```

We wish to see the number of visitors for each museum over a period of time in a single chart. Let's use a lineplot for this.

```python
plt.figure(figsize=(16, 6))
sns.lineplot(data=museum_data)
plt.ylabel('Visitors')
plt.xlabel('Years')
```

We observe a seasonality pattern for the museum Avila Adobe. Let's plot it out separately

```python
plt.figure(figsize=(16, 6))
sns.lineplot(data=museum_data['Avila Adobe'])
plt.xlabel('Year')
plt.ylabel('Visitors')
```

```python
plt.figure(figsize=(16, 6))
sns.boxenplot(data=museum_data)
```

```python
plt.figure(figsize=(16, 6))
sns.scatterplot(data=museum_data)
```

```python
plt.figure(figsize=(16, 6))
sns.boxplot(data=museum_data)
```

```python
plt.figure(figsize=(16, 6))
sns.distplot(museum_data['Avila Adobe'], bins=20)
```

```python
plt.figure(figsize=(16, 6))
sns.distplot(museum_data['Chinese American Museum'], bins=20, color='green', rug=True)
```
