```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
ign_data = pd.read_csv('../datasets/ign_scores.csv', index_col='Platform')
```

```python
ign_data.tail()
```

```python
import numpy as np; np.random.seed(0)
sns.set()
uniform_data = np.random.rand(10, 12)
ax = sns.heatmap(uniform_data, vmin=0, vmax=1)
```

```python
normal_data = np.random.randn(10, 12)
ax = sns.heatmap(normal_data, center=0)
```

```python
flights= sns.load_dataset("flights")
flights = flights.pivot("month", "year", "passengers")
flights.head()
```

```python
plt.figure(figsize=(12, 10))

ax = sns.heatmap(flights, annot=True, fmt="d", linewidths=1, cmap="Reds")
```

```python
plt.figure(figsize=(16, 12))
data = np.random.randn(50, 20)
ax = sns.heatmap(data, xticklabels=2, yticklabels=2, cmap='Blues')
```

```python
f, ax = plt.subplots(figsize=(18, 14))
sns.heatmap(data=ign_data, annot=True, ax=ax, linewidths=0.5, cmap="cubehelix")
```
